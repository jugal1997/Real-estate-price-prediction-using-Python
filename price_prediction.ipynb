```python
## buy_price is our target variable
## understanding the distribution of our target variable

sns.histplot(df_new['buy_price'],kde=True, stat="density", linewidth=0)
```

```python
## From the plot below above we observe, outliers in the column.
sns.boxplot(x=df_new['buy_price'])
```

```python
## Getting rid of the outliers from the 95th percentile
quant=df_new['buy_price'].quantile(0.95)

## Including data that is below the 95th percentile
df1=df_new[df_new['buy_price']<quant]
```

```python
sns.boxplot(x=df1['buy_price'])
## There are much less outliers present than the original distribution
```

```python
## We use log transformation to further normalize our target variable i.e buy_price

log_buy_price=np.log(df1['buy_price'])
df1['log_buy_price']=log_buy_price
df1.head()
```

```python
## one hot encoding of categorical variables
from sklearn.preprocessing import OneHotEncoder
dummy_variables=pd.get_dummies(data=df1,columns=['is_renewal_needed','has_lift','has_parking'],drop_first=True)
```

```python
dummy_variables.head()
```

```python
df_new_=dummy_variables
```

```python
## dropping the old buy_price variable
df_new_.drop(['buy_price'],axis=1,inplace=True)
```

```python
## 
y=df_new_['log_buy_price']
x=df_new_.drop(['log_buy_price'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
x_scaled = scaler.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled,y, test_size=0.3, random_state=42)
print (x_train.shape, y_train.shape)
print (x_test.shape, y_test.shape)
```

```python
## creating a linear regression model
## Training the model
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
y_hat = reg.predict(x_train)
```

```python
## Plotting a scatterplot to visualize the outputs

plt.scatter(y_train, y_hat)
plt.xlabel('Targets (y_train)',size=18)
plt.ylabel('Predictions (y_hat)',size=18)

plt.xlim(10,17)
plt.ylim(10,17)
plt.show()
```

```python
## calculating the r^2 value
r_squared=reg.score(x_train,y_train)
print(r_squared)
```

```python
## Now testing out linear regression model

y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test, alpha=0.2)
plt.xlabel('Targets (y_test)',size=18)
plt.ylabel('Predictions (y_hat_test)',size=18)
plt.xlim(10,17)
plt.ylim(10,17)
plt.show()
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_hat_test, y_test))
print('Mean Squared Error:', metrics.mean_squared_error(y_hat_test, y_test))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_hat_test, y_test)))

from sklearn.metrics import r2_score
r2 = r2_score (y_hat_test, y_test)

print ('R-squared score', round (r2,2))

#adjusted R-squared:
adj_r2 = 1 - (1 - r2) * (254-1)/(254 - 6 - 1)
print ('Adjusted R-squared score', round (adj_r2,2))
```

```python
## Now we manually check the predictions
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Prediction'])
y_test = y_test.reset_index(drop=True)
df_pf['Target'] = np.exp(y_test)
df_pf['Residual'] = df_pf['Target'] - df_pf['Prediction']
df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Target']*100)
df_pf
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
# Finally, we sort by difference in % and manually check the model
df_pf.sort_values(by=['Difference%'], ascending = False)
```

```python
## Training on random forest regressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators = [int(x_scaled) for x_scaled in np.linspace(start = 100, stop = 1000, num = 10)]
#number of features to consider at every split
max_features = ['auto', 'sqrt']
#maximum number of levels in tree
max_depth = [int(x_scaled) for x_scaled in np.linspace(5, 30, num = 6)]
#minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
#minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
## Using random grid search to find best hyperparameters
rf=RandomForestRegressor()
rf=RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', 
                      n_iter = 10, cv = 5, verbose=False, random_state=42, n_jobs = -1)
```

```python
## Training the model
rf.fit(x_train,y_train)
```

```python
rf.best_params_
```

```python
## Making predictions on training data
y_hat_rf = rf.predict(x_train)
```

```python
## Visualizing the outputs
plt.scatter(y_train, y_hat_rf)
plt.xlabel('Targets (y_train)',size=18)
plt.ylabel('Predictions (y_hat_rf)',size=18)
plt.xlim(10,17)
plt.ylim(10,17)
plt.show()
```

```python
## Calculating the r^2 value
r_squared_random_forest=reg.score(x_train,y_train)
print(r_squared_random_forest)
```

```python
plt.scatter(y_test,y_hat_test_rf, alpha = 0.2)
plt.xlabel('Targets (y_test)',size=18)
plt.ylabel('Predictions (y_hat_test_rf)',size=18)
plt.xlim(10,17)
plt.ylim(10,17)
plt.show()
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_hat_test_rf, y_test))
print('Mean Squared Error:', metrics.mean_squared_error(y_hat_test_rf, y_test))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_hat_test_rf, y_test)))

r2_rf = r2_score (y_hat_test_rf, y_test)

print ('R-squared score',  (r2_rf,2))

#adjusted R-squared:
adj_r2_rf = 1 - ((1 - r2_rf) * (254-1)/(254 - 8 - 1))
print ('Adjusted R-squared score',  (adj_r2_rf,2))
```
